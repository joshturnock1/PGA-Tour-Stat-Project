```python
import pandas as pd
```

```python
data = pd.read_csv("../Data Analytics Projects/pgaTourData.csv")
```

```python
data.head()
```

```python
# finding null values
data.isna()
```

# Dropping unnecessary columns

```python
# Dropping Points column as it adds no real value

data.drop(columns=["Points"], inplace=True)

data
```

# Deleting rows with more than 50% of data missing

```python
threshold = len(data.columns) / 2

# Identify rows where the number of missing values is greater than the threshold
rows_with_majority_missing = data.isnull().sum(axis=1) > threshold

# Drop these rows from the DataFrame
data_cleaned = data[~rows_with_majority_missing]

data_cleaned
```

# Replacing missing data

```python
# Calculate the mean of the "Top 10" column
top_10_mean = data_cleaned['Top 10'].mean()

# Define fill values for each column
fill_values = {"Wins": 0, "Top 10": top_10_mean}

# Fill missing values using fillna() with the specified fill values
data_cleaned.fillna(value=fill_values, inplace=True)


data_cleaned
```

```python
# deleting any rows with missing values in "Money" column
data_cleaned.dropna(subset=['Money'], inplace=True)
```

```python
print(data_cleaned.isnull().sum())
```

```python
# save cleaned data to a new CSV file

data_cleaned.to_csv("../Data Analytics Projects/PGA_Tour_cleaned.csv")
```
